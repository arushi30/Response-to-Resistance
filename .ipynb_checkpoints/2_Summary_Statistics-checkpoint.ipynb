```python
import pandas as pd
import numpy as np
import geopandas as gpd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
combined = pd.read_csv("../Personal Response_to_Resistance/Input Data/clean_data.csv", header=0)
#combined = pd.read_csv('clean_data.csv')
combined.head()
```

## Map crime and resistance by census tract

```python
# Get the incident count and resistence count by census tract
resistance_count = pd.DataFrame(combined.groupby("tract")['Target1'].sum()).reset_index()
crime_count = pd.DataFrame(combined.groupby("tract").size()).reset_index()

counts = pd.merge(resistance_count, crime_count, on="tract")
counts.columns = ["TRACT", "resistance_count", "crime_count"]
```

```python
# Read in the 2010 shapefile for the relevant counties 
# Downloaded from https://www.census.gov/geo/maps-data/data/cbf/cbf_tracts.html
map_df = gpd.read_file("Input Data/gz_2010_48_140_00_500k.shp")
map_df = map_df.loc[(map_df["COUNTY"] == "453") | (map_df["COUNTY"] == "491"), ]
```

```python
# Merge the shapefile with the crime data
map_df =pd.merge(map_df, counts, on="TRACT", how="outer", indicator=True)
map_df = map_df.loc[map_df["_merge"]!="left_only", ]
```

```python
vmin, vmax = 0, 70000
fig, ax = plt.subplots(1, figsize=(10, 6))
map_df.plot(column="crime_counts", cmap='Blues', linewidth=0.8, ax=ax, edgecolor='0.8')
sm = plt.cm.ScalarMappable(cmap='Blues', norm=plt.Normalize(vmin=vmin, vmax=vmax))
```
